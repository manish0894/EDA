# 10 least popular songs

```python
sorted_df = df_tracks.sort_values('popularity', ascending= True)
sorted_df.head(10)
```

# 10 most popular songs having popularity > 90

```python
most_popular_songs = df_tracks.query('popularity>90', inplace = False).sort_values('popularity', ascending = False)
most_popular_songs[:10]
```

# Release date as index column

```python
df_tracks.set_index('release_date', inplace = True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms', inplace = True, axis = 1)
```

```python
df_tracks['duration'].head()
```

```python
corr_df = df_tracks.drop(['key','mode','explicit'],axis=1).corr(method = 'pearson')
plt.figure(figsize=(14,6))

heatmap = sns.heatmap(corr_df,annot = True, vmin = -1, vmax = 1, cmap = 'inferno',
                     linewidths = 1, linecolor = 'Black')
heatmap.set_title('Correlation Heatmap between Variable')
# heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

#### Here -1 denotes the variables that have the least or negative correlation, while the values above 0.0 denote the variables with a positive correlation. 



```python
# creating a sample dataset from the full dataset

sample_df = df_tracks.sample(int(0.1*len(df_tracks)))
print(len(sample_df))
```

# Creating a regression plot between Energy & Loudness

```python
plt.figure(figsize = (10,6))
sns.regplot(data=sample_df, y = 'loudness', x = 'energy', color = 'b').set(title='Loudness vs Energy Correlation')
```

#### Analysis: High positive correlation between loudness and energy. data points are in one direction. If the energy increases, the loudness of the song increases and vice-versa

# Regression Plot Between Popularity and Acousticness

```python
plt.figure(figsize = (10,6))
sns.regplot(data=sample_df, y ='popularity',x = 'acousticness', color = 'c').set(title = 'Correlation Plot between Popularity & Acousticness')
```
