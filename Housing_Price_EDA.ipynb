### We see that the original varaiable i.e SalePrice is left-skewed. In such cases, we do either log/fractional-power tranformation of the origina variable to reduce skewness

```python
closeToNormalSP = np.log(df["SalePrice"])
sns.kdeplot(closeToNormalSP)
plt.show()

#Now we can see that log transformed variable is less skewed
```

```python
closeToNormalSP = np.power(df["SalePrice"], 1/4)
sns.kdeplot(closeToNormalSP)
plt.show()

#Now we can see that power transformed variable is less skewed
```
